**PCA is not a full model. but it helps for component reduction**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
type(cancer) # this is not the typical object that we are working with but essentially acts as dictionary
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
cancer['target']
```

```python
cancer['target_names']
```

**it is very difficult to visualize high dimentional data. we use pca to find 2 Principal component. and visualize it into 2 Dimention**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize = (8,6))
plt.grid()
plt.scatter(x_pca[:,0], x_pca[:,1], c= cancer['target'], cmap='plasma')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_comp, cmap='plasma')
```
