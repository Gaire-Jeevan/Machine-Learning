# DIST PLOT

* use to plot one variable
* kde:- kernal density estimation

```python
sns.distplot(tips['total_bill'], kde= False, bins=30)
plt.grid()
```

# Joint plot

* like distplot() but for 2 varable
* kde shows the density where the there is most
* we use jointplot() with default scatter plot inside it

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde') # kind:- reg, hex, kde
```

# Pair plot

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

# RUGPLOT

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
```

# KDE plot

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# Categorical Plot
* barplot
* countplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std) # estimator helps to apply the function to the categorical data
```

# Count plot
* same as barplot but it count the no of data of current axes

```python
sns.countplot(x='sex', data=tips)
```

# BOXPLOT & VIOLIN PLOT
* this type of plot is used to show the distribution of the categorical data

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex',split=True)
```

# STRIPPLOT

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
plt.grid()
```

# SWARMPLOT

```python
sns.violinplot(x='day', y='total_bill', data= tips)
sns.swarmplot(x='day', y='total_bill', data= tips, color='black')
```

# FACTOR PLOT --> catplot()

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

# MATRIX PLOT
* heatmap()
* clustermap()

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot=True,cmap='coolwarm')
```

```python
flights
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white',linewidths=1)
```

## clustermap()
* clustering information based on similarities

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```

# GRIDS

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
plt.grid()

g.map(sns.histplot, 'total_bill',kde=True)
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.scatterplot, 'total_bill','tip')
```

# Regression Plot
* linear model plot (lm plot)

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips, hue='sex',markers=['*','v'], scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips,col='day', row='time', hue='sex', aspect=0.6)
```
