```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

# Functional Method

```python
plt.plot(x, y)
plt.xlabel("X Label")
plt.ylabel("Y label")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

# Object oriented methods

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y label")
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

#axes1.plot(x, y)
axes2.plot(y, x)
axes2.set_title("Smaller Plot")
axes1.set_title("Larger Plot")
```

# FIGURE SIZE & DPI (pixel per inch)

```python
fig = plt.figure(figsize=(8, 2), dpi=100)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))


axes[0].plot(x, y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0,1,1])

ax.plot(x, x**3, label='x Cube')
ax.plot(x,x**2, label='x square')

ax.legend(loc=0) # fit on best location
```

# PLOT APPEARANCE

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='blue', linewidth=3, alpha=0.5, linestyle = '--', marker='o', markersize=20, 
        markeredgecolor='red', markerfacecolor='yellow', markeredgewidth=3) #lw for linewidth
```

# PLOT RANGE

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='r', lw=2, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
