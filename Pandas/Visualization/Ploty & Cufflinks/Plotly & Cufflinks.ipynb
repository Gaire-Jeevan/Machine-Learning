```python
import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected = True) # connect javascript to notebook as plotly essentially connect dataframe and python to into a interactive javascript library
```

```python
cf.go_offline() # that allows to use cufflinks offline
```

```python
# Data

df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({
    'Category' : 'A B C'.split(),
    'Values': [32, 43, 50]
})
```

```python
df2
```

```python
df.iplot()
plt.show()
```

```python
df.iplot(kind='scatter', x='A', y='B',mode='markers', size=10)
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.iplot(kind='bar')
```

```python
df.count().iplot(kind='bar')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```
